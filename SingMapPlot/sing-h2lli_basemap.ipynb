```python
# explanation from GreekDataGuy: https://towardsdatascience.com/lets-plot-airbnb-prices-on-a-map-of-singapore-ddbcae44fb69
# however the site he mentioned to get the map from no longer exist. got mine from https://en.wikipedia.org/wiki/Module:Location_map/data/Singapore
# summary: must know the lat, long limits of the map used; set your plot to those limit too. the rest of the coordinates can google.

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
spore_img   = plt.imread('640px-Singapore_location_map_(main_island).svg.png')
plt.imshow(spore_img, extent=[103.557, 104.131, 1.129, 1.493], alpha=0.8)
plt.ylabel("Latitude", fontsize=12)
plt.xlabel("Longitude", fontsize=12)
plt.ylim(1.129, 1.493)
plt.xlim(103.557, 104.131)

x, y = 103.844, 1.311
flora_x, flora_y = 103.965, 1.357
lli_x, lli_y= 103.893, 1.321
plt.plot(x, y, 'ro', markersize=5, )
plt.text(x, y, ' Singapore', fontsize=8)
plt.plot(flora_x, flora_y, 'bo', markersize=5)
plt.text(flora_x, flora_y, ' 38 Flora Drive', fontsize=8)
plt.plot(lli_x, lli_y, 'bo', markersize=5)
plt.text(lli_x, lli_y, ' Lifelong Learning Institute', fontsize=8)
plt.show()
```
